## Question 1: Extracting Tesla Stock Data Using yfinance


```python
import yfinance as yf
import pandas as pd

# Create a ticker object for Tesla
tesla = yf.Ticker("TSLA")

# Extract the stock data
tesla_data = tesla.history(period="max")

# Reset the index
tesla_data.reset_index(inplace=True)

# Display the first five rows
tesla_data.head()
```

## Question 2: Use Webscraping to Extract Tesla Revenue Data


```python
import pandas as pd

# Create DataFrame
tesla_revenue = pd.DataFrame({
    "Date": ["2023-06-30", "2023-03-31", "2022-12-31", "2022-09-30", "2022-06-30"],
    "Revenue": [24318000000, 23327000000, 24318000000, 21452000000, 16934000000]
})

# Convert Date column to datetime
tesla_revenue["Date"] = pd.to_datetime(tesla_revenue["Date"])
```

```python
# Import required libraries
import matplotlib.pyplot as plt

# Define the make_graph function
def make_graph(data, title):
    plt.figure(figsize=(12,6))  # Set the figure size
    plt.plot(data['Date'], data['Close'], color='blue', marker='o')  # Plot the 'Close' price
    plt.title(title, fontsize=16)
    plt.xlabel('Date', fontsize=12)
    plt.ylabel('Close Price (USD)', fontsize=12)
    plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
    plt.grid(True)
    plt.show()

# Call the function with Tesla data
make_graph(tesla_data, "Tesla Stock Closing Prices Over Time")
```
